```python
# Read POSTS raw data
df_posts = spark.read.json("abfss://bronze@stlogdatadivya.dfs.core.windows.net/raw/posts/")
```

```python
from pyspark.sql.functions import current_timestamp
#Add ingestion timestamp
df_posts = df_posts.withColumn("ingestion_time", current_timestamp())
```

```python
#Write POSTS to Bronze Delta Table
df_posts.write.format("delta") \
    .mode("append") \
    .option("mergeSchema", "true") \
    .saveAsTable("log_analytics.bronze.posts_raw")
```

```python
#Read COMMENTS raw data
df_comments = spark.read.json("abfss://bronze@stlogdatadivya.dfs.core.windows.net/raw/comment/")
```

```python
#Add ingestion timestamp
df_comments = df_comments.withColumn("ingestion_time", current_timestamp())
```

```python
#Write COMMENTS to Bronze Delta Table
df_comments.write.format("delta") \
    .mode("append") \
    .option("mergeSchema", "true") \
    .saveAsTable("log_analytics.bronze.comments_raw")
```
